```python
# Matplotlib is a data visualization library in Python that produces publication-quality plots.
# It integrates well with NumPy (for numerical computations) and Pandas (for structured data analysis).
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate 100 points between 0 and 10

# linspace() function in NumPy returns an array of 
# evenly spaced numbers over a specified range. Unlike the range() function in Python 
# that generates numbers with a specific step size. linspace() allows 
# user to specify the total number of points you want in the array, 
# and NumPy will calculate the spacing between the numbers automatically.
x = np.linspace(0,10, 10)
print(x)
#plotting sine graph function
y = np.sin(x)

#plot grid
plt.grid(True,linestyle="--",linewidth=0.5)
plt.gca().set_facecolor("lightblue")

# Create a plot
# plt.plot(x,y, label="sin(x)", color="orange", linestyle="--")
#apply alpha for transparency
# You can use the keyword argument marker to emphasize each point with a specified marker
plt.plot(x,y, label="sin(x)", linestyle="dashdot",marker="o", color="orange", alpha=0.8)

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Test plot")

# Show legend
plt.legend()

# Display plot
plt.plot()
```

# Lineplot

```python
#A line plot shows data points connected by a line, typically used to visualize trends over a continuous variable like time.

#points
x = np.array([20,30,40,50,60])
y = np.array([25,26,28,30,52])

#plotting
plt.plot(x,y)
plt.title("Line Plot Example")
plt.xlabel("X axis")
plt.ylabel("Y axis")

#legend
# plt.legend()

#show
plt.show()
```

# ScatterPlot

```python
#A scatter plot is used to visualize the relationship between two numerical variables. Each point represents a pair of values.

#points
x = np.array([22,30,42,50,60])
y = np.array([25,26,28,30,52])

#plotting 
plt.scatter(x, y, color='r')
#annotate like add extra information to a plot, like pointing out a specific point on a graph
#Imagine you are drawing a picture and you want to add a note next to something important.
# plt.annotate("Imp point this is demo",xy=(3, 6), xytext=(4,5),arrowprops=dict(facecolor='blue', shrink=0.05))
plt.title("ScatterPlot example")
plt.xlabel("X")
plt.ylabel("Y")

#show
plt.show()
```

# BarChart

```python
# A bar chart is used to compare quantities of different categories. Each category is represented by a bar.

#points
x = np.array([22,30,42,50,60])
y = np.array([27,26,28,30,52])

#plotting bar stand for vertical barh stands for horizontal
plt.bar(x, y, color = 'green',width=2)
plt.title("Barchart title")
plt.xlabel("X")
plt.ylabel("Y")

#show
plt.show()
```

# Histogram

```python
# A histogram is used to show the distribution of data. It groups data points into bins and displays the frequency of each bin.
data = np.array([12,23,44,21,41,87,90,32,77])

#plot
plt.hist(data, bins=5, color='lightblue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Bins")
plt.ylabel("Freq")

#show
plt.show()
```

# Subplot

```python
# With the subplot() function you can draw multiple plots in one figure

x = np.array([0,1,2,4])
y = np.array([3,8,1,10])

#subplot takes 3 arguments e.g. (1,2,1) 1 row, 2 columns, and this plot is the first plot
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Subplot 1")
plt.xlabel("X")
plt.ylabel("Y")

x = np.array([10,12,14,22,26])
y = np.array([12,14,16,18,20])

#subplot takes 3 arguments e.g. (1,2,2) 1 row, 2 columns, and this plot is the second plot
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("Subplot 2")
plt.xlabel("X")
plt.ylabel("Y")


# if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side
# x = np.array([11,12,14,22,26])
# y = np.array([15,14,16,18,20])

# plt.subplot(2,1,1)
# plt.plot(x,y)
# plt.title("Subplot 1")
# plt.xlabel("X")
# plt.ylabel("Y")

# x = np.array([11,12,14,22,26])
# y = np.array([15,14,16,18,20])

# plt.subplots_adjust(top=2.9)

# plt.subplot(2,1,2)
# plt.plot(x,y)
# plt.title("Subplot 2")
# plt.xlabel("X")
# plt.ylabel("Y")

#adding super title
plt.suptitle("Subplots")

plt.show()
```

# Pie Charts

```python
labels = np.array(["Apple", "Banana", "Cherry", "Date"])
values = np.array([30,40,50,60])
#The startangle parameter is defined with an angle in degrees, default angle is 0 starts from right side in anti clockwise direction
plt.pie(values, labels = labels, startangle=90, shadow=True)
plt.title("Pie Chart")

plt.legend()
plt.show()
```

# Heatmap

```python
user_data = np.array([[3,4,5], [6,7,8], [9,12,13]])

#plotting heatmap
plt.imshow(user_data, cmap='coolwarm', interpolation='nearest')
plt.title("Heatmap sample")
plt.ylabel("Y")
plt.xlabel("X")

plt.colorbar()

plt.show()
```
