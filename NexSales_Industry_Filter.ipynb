```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('final_dataset.csv', low_memory=False)

#set max columns and rows as none to get all values
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df.head(5)
```

```python
#Creating new columns to get refined industry values  
df["Refined_Industry"] = df["Industry"]
df.sample(5)
```

```python
#get value counts of column frequency
df["Refined_Industry"].value_counts()
```

```python
#replace linkedin values with nan value
df.loc[df["Refined_Industry"].str.contains('https://www.linkedin.com/company', na=False)] = np.nan

df["Refined_Industry"].value_counts()
```

```python
df.sample(5)
```

```python
# get all columns info
df.info()
```

```python
# replace na with company industry values 
df["Refined_Industry"].fillna(df["Company Industry"], inplace=True)
```

```python
# get refined values count with no nan values
df["Refined_Industry"].value_counts()
```

```python
#check if na values are removed or not
df["Refined_Industry"].isna().sum()
```

```python
#replace unknown industry values with others
df.loc[df["Refined_Industry"].str.contains("Electricity, Oil & Gas", na=False)] = "Electricity, Oil & Gas"
df.loc[df["Refined_Industry"].str.contains("Hospital|Healthcare", na=False)] = "Hospitality"
df.loc[df["Refined_Industry"].str.contains("Transport", na=False)] = "Transportation"
df.loc[df["Refined_Industry"].str.contains("Manufacturing", na=False)] = "Manufacturing"
df.loc[df["Refined_Industry"].str.contains("Company|\\$1.9B|NV|Companies|SUB|Services|Local|Organizations", na=False)] = "Others"

df["Refined_Industry"].value_counts()
```
