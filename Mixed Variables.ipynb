```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
bar_graph = df['number'].value_counts().plot.bar()
bar_graph.set_title("No. of Passengers travelling in groups")
```

```python
# extract numerical data
df['number_numer'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')

# extract categorical data
df['number_categ'] = np.where(df['numerical_number'].isnull(), df['number'], np.nan)

df.head()
```

```python
#splitting numerical and categorical value from Cabin variable r stands for raw value
df['cabin_num'] = df['Cabin'].str.extract(r'(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]

df.head()
```

```python
#plotting bar graph for cabin category passengers
df['cabin_cat'].value_counts().plot.bar()
```

```python
#extracting ticket number and categorical value

#extracts all number values after space in ticket
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])

#extracts alphabetical values
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])

df.head(10)
```
