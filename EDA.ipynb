# Univariate Analysis

### Single column analysis as independent variable is called as univariate analysis
### Two column analysis as independent variable is called as bivariate analysis
### Two or more column analysis as independent variable is called as multivariate analysis
### Check whether data is numerical or categorical data before start analysis

## Numerical Data

## Histogram

```python
# better use with numerical data for understanding distribution it also has kde
# KDE stands for Kernel Density Estimate. It is like drawing a smooth curve over the bars of the histogram to show the shape of the data.
#ðŸ“Œ Think of it like spreading peanut butter smoothly over a sandwich. Instead of sharp jumps (bars), KDE smooths the shape of the histogram.
sns.histplot(data=df['Age'], bins=10) 
```

## Distplot

```python
#better version of histogram with density line
sns.displot(df['Age'], bins=5, kde=True)
```

## BoxPlot

```python
# Imagine you have a box full of chocolates ðŸ«, and you want to divide them fairly among your friends. Some people get more, some get less, 
# but you want to see the distribution clearly. A Boxplot (also called a box-and-whisker plot) helps us visualize 
# the spread of dataâ€”showing minimum, maximum, median, and outliers.

# Understanding a Boxplot: Chocolate ðŸ« Example
# ðŸ”¹ Think of a chocolate box

# The box in the middle shows the main group of chocolates that most people get.
# The line inside the box is the median (the middle point).
# The "whiskers" (lines outside the box) show the smallest and largest numbers.
# Any chocolates that are far away (outliers) are students who got very high or very low scores.

sns.boxplot(x=df['Pclass'], y = df['Age'], color='red')
```

# Bi-variate and multivariate analysis

## Scatterplot (Numerical - Numerical)

```python
restaurant_bills = sns.load_dataset("tips")

healthexp = sns.load_dataset("healthexp")

flights = sns.load_dataset("flights")

attention = sns.load_dataset("attention")
```

```python
restaurant_bills
```

```python
healthexp
```

```python
flights
```

```python
attention
```

```python
sns.scatterplot(x=restaurant_bills['total_bill'],y=restaurant_bills['tip'], hue=restaurant_bills['day'], style=restaurant_bills['smoker'], size=restaurant_bills['size'])
```

## BarPlot (Numerical - Categorical)

```python
sns.barplot(x=healthexp["Country"], y=healthexp["Spending_USD"])
```

```python
sns.barplot(x=healthexp["Country"], y=healthexp["Spending_USD"], hue=healthexp["Year"], palette='dark:green')
```

```python
sns.barplot(x=healthexp["Country"], y=healthexp["Spending_USD"], hue=healthexp["Life_Expectancy"], palette='dark:red')
```

## BoxPlot (Numerical - Categorical)

```python
sns.boxplot(x = attention["attention"], y = attention["subject"], hue=attention["solutions"])
```

```python
sns.boxplot(x = attention["attention"], y = attention["subject"], hue=attention["score"])
```

## DistPlot (Numerical - Categorical)

```python
sns.displot(flights["passengers"], bins = 10, kde=True)
```

```python
sns.displot(data = flights, y = "passengers",x= "month", bins = 5, hue="year")
```

## Heatmap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(restaurant_bills['smoker'],restaurant_bills['day']))
```

## ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(healthexp["Life_Expectancy"], healthexp["Country"]))
```

## PairPlot (Numerical - Numerical)

```python
sns.pairplot(healthexp)
```
